```python
import string
import re
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
synopses = test_df['all_text_2'].tolist()
```

```python
pkl_filename_master = "stop_words.pkl"
with open(pkl_filename_master, 'rb') as picklefile: 
    stop_words= pickle.load(picklefile)
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

def tokenize_and_stem(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word for sent in nltk.sent_tokenize(str(text)) for word in nltk.word_tokenize(sent)]
    tokens = [s.encode('ascii','ignore') for s in tokens] 
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(str(token))
    stems = [str(stemmer.stem(t)) for t in filtered_tokens]
    stems  = [s.encode('ascii','ignore') for s in stems ] 
    #print(stems)
    return stems
```

```python
tokens = tokenize_and_stem(synopses);
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

#define vectorizer parameters
tfidf_vectorizer = TfidfVectorizer(max_df=0.5, max_features=2000,
                                 min_df=10, stop_words=stop_words,
                                 use_idf=True, tokenizer= tokenize_and_stem, ngram_range=(1,3))

%time tfidf_matrix = tfidf_vectorizer.fit_transform(synopses) #fit the vectorizer to synopses

print(tfidf_matrix.shape)
```

```python
test_df['tokens'] = 1
```

```python
feature_names = tfidf_vectorizer.get_feature_names()

for n in range(1600):
    word_list = []
    for col in tfidf_matrix[n].nonzero()[1]:
        word_list.append(feature_names[col])
        #print(feature_names[col], ' - ', tfidf_matrix[0, col])
    #print(word_list)
    test_df['tokens'][n] = word_list
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
test_df['index1'] = test_df.index
```

```python
test_df.head(3)
```

```python
rows = range(len(test_df))
columns = ['name','recommendations']

df_rec = pd.DataFrame(index=rows, columns=columns)

recommendations = []
for idx, row in test_df.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-60:-1]
    similar_items = [(cosine_similarities[idx][i], test_df['name'][i],test_df['geek_rating'][i],cosine_similarities[idx][i]*test_df['geek_rating'][i] )
                     for i in similar_indices]
    flattened = similar_items[1:]
    flattened_list = [list(x) for x in flattened]
    flattened_list = [list(x) for x in flattened_list]
    df_rec['name'][idx] = test_df['name'][idx]
    df_rec['recommendations'][idx] = flattened_list

df_rec.head(10)
    
```

```python
df_stacked_rec = df_rec.set_index('name')['recommendations'].apply(pd.Series).stack()
df_stacked_rec = df_stacked_rec.reset_index()
df_stacked_rec.columns = ['name','Similarity_rank','recommendation']
df_stacked_rec.head(10)
```

```python
df_rec_details = df_stacked_rec['recommendation'].apply(pd.Series)
#df_rec_details = df_rec_details.rename(columns = ['similarity_score','Name','Geek_score','Total_score'])
df_stacked_rec = pd.concat([df_stacked_rec[:],df_rec_details[:]],axis = 1)
df_stacked_rec = df_stacked_rec.rename(columns={0:'similarity_score',1:'Game Name',2:"Geek_rating",3:'Total_score'})
df_stacked_rec.head(20)
```

```python
df_stacked_rec_grouped = df_stacked_rec.groupby('name')
```

```python
def Game_recommend(game):
    recommendation_list = df_stacked_rec[df_stacked_rec['name'] == game]
    Sorted_recommendation_list = recommendation_list.sort_values(by = 'Total_score', ascending=[False])
    Recommendation_table = Sorted_recommendation_list.drop(['name','Similarity_rank','recommendation','Geek_rating','similarity_score','Total_score'],1)
    
    print("You may also like these games:")
    return(Recommendation_table)
```
